```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv('datagym_customers.csv')
visits = pd.read_csv('datagym_visits.csv')
```

```python
customers[:5]
```

```python
visits[:5]
```

```python
visit2 = visits.groupby('customer_id').count()
```

```python
visit2[:3]
```

```python
visit2 = visit2.reset_index()
```

```python
customersVisit = pd.merge(customers, visit2, how ='outer')
```

```python
customersVisit[:5]
```

```python
customersVisit.rename(columns={
    'date': 'noOfVisits',
}, inplace=True)
```

```python
customersVisit[:5]
```

```python
customersVisit['avgVisits'] = customersVisit['noOfVisits']/customersVisit['duration']
```

```python
customersVisit.head()
```

```python
customersVisit['quit'] = customersVisit['date_quit'].isnull()
```

```python
customersVisit.head()
```

```python
customersVisit.date_quit.max()
```

```python
customersVisit['duration_all'] = customersVisit.date_quit.replace('2015-06-29', np.NaN, inplace=True)
```

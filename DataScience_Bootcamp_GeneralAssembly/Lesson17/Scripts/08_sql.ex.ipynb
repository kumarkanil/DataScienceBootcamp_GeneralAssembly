```python
# Open connection to database
# (Download from http://chinookdatabase.codeplex.com/)
db_connection = sqlite3.connect('../../Data/Chinook_Sqlite.sqlite')
```

```python
# Select the first 10 records from 'customer'
pd.io.sql.read_sql('SELECT * FROM customer LIMIT 10', con=db_connection)
```

```python
pd.io.sql.read_sql('SELECT DISTINCT Country FROM Customer', con=db_connection)
```

```python
# Select the first name of all customers from the UK
pd.io.sql.read_sql('SELECT FirstName FROM customer WHERE Country == "United Kingdom"', con=db_connection)
```

```python
# Select the city and country of all customers from the UK or Portugal
pd.io.sql.read_sql('SELECT City, Country FROM customer WHERE Country == "United Kingdom" OR Country == "Portugal"', con=db_connection)
```

```python
# Select the first 10 records from 'invoice'
pd.io.sql.read_sql('SELECT * FROM Invoice LIMIT 10', con=db_connection)
```

```python
# Join 'customer' and 'invoice', and retrieve customer ID and invoice amount
pd.io.sql.read_sql('SELECT customer.CustomerId, Invoice.Total FROM customer INNER JOIN Invoice ON customer.CustomerId = Invoice.InvoiceId', con=db_connection)
```

```python
# (Continued) Compute the total of all invoices by customer
pd.io.sql.read_sql('SELECT SUM(Total) FROM Invoice', con=db_connection)
```

```python
# (Continued) Aggregate only invoices from 2013
pd.io.sql.read_sql('''SELECT SUM(Total) FROM Invoice WHERE STRFTIME(\'%Y\', Day) == \'2013\'''', con=db_connection)
```
